### Prepare the data and ensure data types are correct

## Calculate monthly average precipitation against reference period 1971-2000

To calculate the precipitation anomalies using a reference period of 1971-2000, you would follow these steps:
1. Calculate Reference Precipitation (pr_ref)
a) Filter Data for Reference Period (1971-2000):

    Filter your dataset to include only data from the years 1971 to 2000.
    This subset of data will be used to establish the "normal" precipitation levels against which anomalies will be measured.

b) Calculate Average Monthly Precipitation:

    Group this subset by month (regardless of the year) and calculate the average precipitation for each month.
    This gives you the average monthly precipitation for the reference period. Each month (January, February, etc.) will have its own average value based on data from 1971-2000.

2. Join pr_ref to Main Data
a) Create pr_ref Dataset:

    The dataset you created in step 1, which contains the average monthly precipitation for each month (from January to December) over the years 1971-2000, is your pr_ref dataset.

b) Join pr_ref to Your Main Dataset:

    Perform a left_join operation to add the pr_ref data to your main dataset.
    Join on the month so that each record in your main dataset gets the average precipitation value for its respective month from the pr_ref dataset.

3. Calculate Anomalies
a) Anomaly Calculation:

    For each record in your main dataset, calculate the anomaly as the percentage difference between the actual precipitation (prs) and the reference precipitation (pr_ref) for that month.
    The formula ((prs*100/pr_ref)-100) calculates this percentage deviation. Subtracting 100 sets the average (i.e., no anomaly) to 0%.

b) Positive or Negative Anomaly:

    Create a variable (sign) to indicate whether the anomaly is positive or negative. This can be done using a conditional statement (e.g., ifelse(anom > 0, "positive", "negative")).

```python
# Step 2: Calculate Reference Precipitation

ref_period = df[(df['Year'] >= 1971) & (df['Year'] <= 2000)]
pr_ref = ref_period.groupby('Month')['Mean'].mean().reset_index()

pr_ref
```

```python
# Step 3: Join pr_ref to Main Data
df = df.merge(pr_ref, on='Month', suffixes=('', '_ref'))

df.head()
```

```python
# Step 4: Calculate Anomalies
df['Anomaly'] = ((df['Mean'] / df['Mean_ref']) - 1) * 100 # in %
```

```python
df.head()
```

```python
# Step 5: Additional Variable "sign" referring to positive or negative anamoly
df['Sign'] = df['Anomaly'].apply(lambda x: 'positive' if x > 0 else 'negative')

# Display the updated dataframe
df
```

### Test Plot with one year

```python
# Filter data for the year 2018
df_2018 = df[df['Year'] == 2018]

# Set the style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Date', y='Anomaly', hue='Sign', data=df_2018, palette=['#99000d', '#034e7b'])

# Customize the plot
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Precipitation anomaly (%)')
plt.title('Monthly Precipitation Anomalies for 2018')
plt.legend(title='Anomaly', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust x-axis to show each month
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=12))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group by month and calculate the statistical metrics
data_norm = df.groupby('Month').agg(
    mx=pd.NamedAgg(column='Anomaly', aggfunc='max'),
    min=pd.NamedAgg(column='Anomaly', aggfunc='min'),
    q25=pd.NamedAgg(column='Anomaly', aggfunc=lambda x: x.quantile(0.25)),
    q75=pd.NamedAgg(column='Anomaly', aggfunc=lambda x: x.quantile(0.75))
)

# Calculate the Interquartile Range (IQR)
data_norm['iqr'] = data_norm['q75'] - data_norm['q25']
data_norm
```

    max for the maximum value of anomalies.
    min for the minimum value of anomalies.
    quantile(0.25) for the 25th percentile.
    quantile(0.75) for the 75th percentile.
