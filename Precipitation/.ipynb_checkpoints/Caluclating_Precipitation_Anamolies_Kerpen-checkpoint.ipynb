# Calculating climatologies and anomalies

```python
# Read in monthly average precipiation data over Kerpen

df = pd.read_csv(r"../data/monthly_p_kerpen_by_month_1951-2022.csv")
df.head()
```

```python
# Basic overview of descriptive stats
df.describe()
```

## Calculate monthly average precipitation

To calculate the average precipitation for each month across all years from your DataFrame, you can follow these steps:

   - Ensure that the "Date" column is recognized as a datetime object.
   - Extract the month from the "Date" column.
   - Group the data by the extracted month.
   - Calculate the mean precipitation for each month.

```python
# Ensure "Date" column is recognized as datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract the month from the "Date" column and create a new column "Month"
df['Month'] = df['Date'].dt.month

# Group by month and calculate the mean precipitation for each month
monthly_mean_precipitation = df.groupby('Month')['Mean'].mean().reset_index()

# Rename the "Mean" column to "LongTermMean"
monthly_mean_precipitation.rename(columns={'Mean': 'LongTermMean'}, inplace=True)


# Display the resulting DataFrame
monthly_mean_precipitation.head(12)
```

```python
#save the results of the df to a .csv
df.to_csv(r"../results/monthly_avg_p_kerpen_1951-2022.csv", index=False, header=True, encoding='utf-8')
print(".csv has been created")
```

## Calculating the standard deviation

We will apply `standard deviation = X(observed value) - monthly_mean_precipitation`

```python
# This is what our df looks like

df.head()
```

To calculate anomalies (deviations from the long-term average) for each monthly mean value in our df, we can use the monthly_mean_precipitation DataFrame, which contains the long-term average monthly values.

```python
# Extract the month from the "Date" column and create a new column "Month"
df['Month'] = df['Date'].dt.month
df.head()
```

```python
# Merge the long-term average values from "observed_monthly_means" into the df based on the "Month" column
df = df.merge(monthly_mean_precipitation, on='Month', suffixes=('', '_LongTermAvg'))
```

```python
# Calculate the anomaly (deviation from long-term average)
df['Std_Deviation'] = df['Mean'] - df['LongTermMean']

df
```

```python
df.describe()
```

```python
# Visualize deviations from the reference mean

# Create a bar chart
fig = px.bar(df, x='Date', y='Std_Deviation',
             title='Temperature Anomalies Over Time',
             labels={'Std_Deviation': 'Precipitation (mm)', 'Date': 'Date'})

# Update X and Y axis labels
fig.update_xaxes(showticklabels=False, tickformat='%b %Y', title_text='')
fig.update_yaxes(dtick=15, title_text='Deviation from Longterm Precipitation (mm)')


# Assign colors based on the sign of the anomaly
fig.update_traces(marker_color=df['Std_Deviation'].apply(lambda x: 'blue' if x > 0 else 'red'))

# Display the figure
fig.show()
```

## Visualize monthly precipitation anomalies

based on https://dominicroye.github.io/en/2019/visualize-monthly-precipitation-anomalies/

```python
df_2020 = df[df['Date'].dt.year == 2020]


# Extract year and month from the Date column
df_2020['Year'] = df_2020['Date'].dt.year
df_2020['MonthName'] = df_2020['Date'].dt.strftime('%b')

# Create a pivot table for the heatmap
heatmap_data = df_2020.pivot_table(values='Std_Deviation', index='MonthName', columns='Year', aggfunc='mean')

# Create a color palette (you can customize the colors)
colors = sns.color_palette("coolwarm", as_cmap=True)

# Create the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap=colors, annot=True, fmt=".1f", linewidths=0.5, cbar_kws={'label': 'Standard Deviation'})
plt.title('Precipitation Standard Deviation Calendar Heatmap (1951-2022)')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```

```python
def annual_calendar(year):
    nrow = 3
    ncol = 4
    figsize = (10, 6)
    fig, axs = plt.subplots(figsize=figsize, nrows=nrow, ncols=ncol)

    month = 1

    for ax in axs.reshape(-1):
        ax.axis([0, 7, 0, 7])
        ax.set_axis_off()
        month_name = calendar.month_name[month]
        ax.text(3, 3, month_name, ha='center', va='center', fontsize=12)
        month += 1

annual_calendar(2020)
plt.show()
```
